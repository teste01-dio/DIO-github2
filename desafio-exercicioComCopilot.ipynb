```python
# 1) veririficar se número é par ou impar 
def verificaPar (numero):
    if numero % 2 == 0:
        return "O número é par"
    else:
        return "O número é ímpar"
    
```

```python
verificaPar(int(input("Digite um número: ")))
```

```python
# 2) def calcular a média de N números

def calcularMedia(lista):
    return sum(lista) / len(lista)

# digitar uma lista de números separados por espaço e decimais utilizando '.'

numeros = list(map(float, input("Digite uma lista de números separados por espaço: ").split()))

print(calcularMedia(numeros))
```

```python
# 3) verificar palíndromos
# Um palíndromo é uma palavra ou frase que pode ser lida de duas formas
# diferentes, da mesma maneira. Ex: "ovo", "arara", "asa", "radar", "reviver"


def verificaPalindromo(palavra):
    if palavra == palavra[::-1]:
        return "A palavra é um palíndromo"
    else:
        return "A palavra não é um palíndromo"
    
verificaPalindromo(input("Digite uma palavra: "))

```
